```python
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from PIL import Image as im
from pathlib import Path
```

```python
for i in Path('datasets/med-image/testB/').iterdir():
    _, axarr = plt.subplots(1, 3)
    
    orig = im.open(i)
    axarr[0].imshow(orig)
    orig = np.asarray(orig, dtype=np.uint8)
    
    orig = orig[:,:,0].flatten()
    plt.figure(figsize=(8,8))
    axarr[1].hist(orig, bins=20)
    
    orig = orig>120
    axarr[2].imshow(orig.reshape(800, 800))
    plt.show()
```
